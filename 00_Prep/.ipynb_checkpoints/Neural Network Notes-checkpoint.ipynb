## 6. Perceptron
how a simple neural network makes decisions: 
 - taking in input data 
 - processing that information
 - finally, producing an output in the form of a decision

### 6.1 Weights
when we initialize a neural network, we don't know what information will be most important in making a decision. It's up to the neural network to learn for itself which data is most important and adjust how it considers that data.

When input data comes into a perceptron, it gets multiplied by a weight value that is assigned to this particular input.

### 6.2 Linear Combination
$$
\large
w_{grade} \times x_{grade} + w_{test} \times x_{test} = \sum_{n=1}^{m} w_{i}x_{i}
$$

### 6.3 Calculating the Output with an Activation Function
Finally, the result of the perceptron's summation is turned into an output signal! This is done by feeding the linear combination into an activation function.

Complete perceptron Formula using heaviside step function as activation function:

$$
f(x_{1}, x_{2},...,x_{m}) = \left\{
                \begin{array}{ll}
                  0 \hspace{5mm} if \hspace{2mm} b + \sum_{i=1}^{i=m} w_{i}x_{i} < 0\\
                  1 \hspace{5mm} otherwise
                \end{array}
              \right.
$$

Then the neural network starts to learn! Initially, the weights ($w_{i}$) and bias (b) are assigned a random value, and then they are updated using a learning algorithm such as $\Large Gradient \hspace{1mm} Descent$

```python
import pandas as pd

# TODO: Set weight1, weight2, and bias
weight1 = 0.15
weight2 = 0.15
bias = -0.2


# DON'T CHANGE ANYTHING BELOW
# Inputs and outputs
test_inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
correct_outputs = [False, False, False, True]
outputs = []

# Generate and check output
for test_input, correct_output in zip(test_inputs, correct_outputs):
    confidence = weight1 * test_input[0] + weight2 * test_input[1] + bias
    output = int(confidence > 0)
    is_correct_string = 'Yes' if output == correct_output else 'No'
    outputs.append([test_input[0], test_input[1], confidence, output, is_correct_string])

# Print output
num_wrong = len([output[4] for output in outputs if output[4] == 'No'])
output_frame = pd.DataFrame(outputs, columns=['Input 1', '  Input 2', '  Confidence', '  Activation Output', '  Is Correct'])
if not num_wrong:
    print('Nice!  You got it all correct.\n')
else:
    print('You got {} wrong.  Keep trying!\n'.format(num_wrong))
print(output_frame.to_string(index=False))
```

$$ \large

& w_{1} + w_{2} + \beta > 0 \\
& w_{1} + \beta < 0 \\

& \beta < 0
\end{align}
$$

## 7 Sigmoid function as the activation
$$
\large
sigmoid(x) = \frac {1} {1 + e^{-x}}
$$
Sigmoid function is bounded between 0 and 1

Once start using activation functions that are continuous and differentiable, it's possible to train the network using gradient descent

### Simple Neural Network Implementation
Use numpy to calculate the output of a simple network with two input nodes and one output node with a sigmoid activation function.

- Calculate the output of the network

```python
import numpy as np
import pandas as pd
```

## Gradient Descent
- Need to learn the weights from example data, then use those weights to make the predictions
- We want the network to make predictions as close as possible to the real values. To measure this, we need a metric of how wrong the predictions are, the error. A common metric is the sum of the squared errors (SSE):
$$
SSE = \frac{1}{2} \sum_{\mu}\sum_{j} \left(y_{j}^{\mu} - \hat y_{j}^{\mu}\right)^{2}
$$

The SSE is a good choice for a few reasons. The square ensures the error is always positive and larger errors are penalized more than smaller errors.

The output of a neural network, the prediction, depends on the weights
$$
\hat y_{j}^{u} = f \left(\sum_{i}w_{ij}x_{i}^{\mu}\right)
$$

As a result the error depends on the weights:
$$
SSE = \frac{1}{2} \sum_{\mu}\sum_{j} \left(y_{j}^{\mu} - f \left(\sum_{i}w_{ij}x_{i}^{\mu}\right)\right)^{2}
$$

Our goal is to find weights $w_{ij}$ that minimize the SSE, to do this with a neural netowrk, typically will use  Gradient Descent

Gradient is another term for rate of change or slope. To calculate a rate of change, we turn to calculus, specifically derivatives. A derivative of a function $f(x)$ gives you another function $f'(x)$ that returns the slope of $f(x)$ at point x.

The gradient is just a derivative generalized to functions with more than one variable. 

### Math For Gradient Descent
1. Calculate the gradient of the error: need the partial derivatives of the error with respect to each of the weight
$$
SSE = \frac{1}{2} \sum_{j} \left(y_{j} - \hat y_{j}\right)^{2}
$$

2. The Prediction $\hat y_{j}$ is
$$
\hat y_{j} = f \left(\sum_{i}w_{ij}x_{i}\right)
$$

3. Update weights by an amount $\Delta w_{ij}$ proportional to the gradient of the error but in the opposite direction:
$$
\Delta w_{ij} = - \eta\frac{\partial E}{\partial w_{ij}}
$$
$\eta$ is the learning rate which allows us to scale the step size in the gradient descent updates

4. Now understand what $\frac {\partial E} {\partial w_{ij}}$ will be, where $E = \frac {1} {2} \sum_{j}\left(y_{j} - \hat y_{j}\right) ^ {2}$

$$\large
\begin{align}
\frac {\partial E}{\partial w_{ij}} & = (y_{j} - \hat y_{j}) \frac {\partial}{\partial w_{ij}}(y_{j} - \hat y{j}) \\
& = -(y_{j}-\hat y_{j}) \frac {\partial \hat y_j}{\partial w_{ij}} \\
& = -(y_{j}-\hat y_{j}) f'(\sum_{i}w_{ij}x_{i}) \frac {\partial}{\partial w_{ij}} \sum_{i}w_{ij}x_{i} \\
& = -(y_{j}-\hat y_{j}) f'(\sum_{i}w_{ij}x_{i}) x_{i} \\
& where, h_{j} = \sum_{i}w_{ij}x_{i}
\end{align}
$$

So:
$$\large
\begin{align}
\frac {\partial E}{\partial w_{ij}} & = -(y_{j}-\hat y_{j}) f'(h_{j}) x_{i} \\
\Delta w_{ij} & = \eta (y_{j}-\hat y_{j})f'(h_{j}) x_{i}
\end{align}
$$

#### Intuition:
- $(y_{j} - \hat y_{j})$ is the prediction error. The larger this error is, the larger the step should be.
- The input term $x_{i}$, larger inputs drive more error, we need to scale the weight proportional to input as well
- Gradient $f'(h_{j})$ , This is proportional to the effect unit $j$ has on the output. Remember that the gradient is a rate of change. If the gradient is small, then a change in the unit input $h_{j}$ will have a small effect on the error. And conversely, if the gradient is large, a change in the unit input will have a large effect. This term produces larger gradient descent steps for units that have larger gradients and therefore effect the change in error more.

If we define the error gradient $\delta_{j}$ as:
$$\large
\delta_{j} = (y_{j}-\hat y_{j})f'(h_{j})
$$
then we can write the gradient descent step as:

$$\large
\Delta w_{ij} = \eta \delta_{j} x_{i}
$$

### Gradient Descent Exercise
- calculate one gradient descent step for the weights of a simple network with two inputs and one output unit.
- for sigmoid function it's derivative can be written in terms of the sigmoid function itself:
$$
f'(h) = f(h) (1-f(h))
$$

```python
import numpy as np
```

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))
```

```python
def run():
    learn_rate = 0.5
    x = np.array([1, 2])
    y = np.array([0.5])
    
    #Initial rate
    w = np.array([0.5, -0.5])
    
    ## Calculate One Gradient Descent Step for Each Weight
    #0 Calcualte the linear sum
    nn_h = np.dot(x, w)
    #1 Calculate the output of neural network
    nn_output = sigmoid(nn_h)
    print(nn_output)
    #2 Calculate the error of output
    nn_error = nn_output - y
    print(nn_error)
    #3 Calculate the derivative of activation function
    nn_f_d = sigmoid(nn_h) * (1-sigmoid(nn_h))
    print(nn_f_d)
    #4 Calculate change in weights
    delta_w = -learn_rate * nn_error * nn_f_d * x
    print(delta_w)
```

```python
#Write the main function
if __name__ == "__main__":
    run()
    
```
